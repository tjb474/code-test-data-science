# Read in JSON and create features 

```python
path = "..\\code-test-data-science\\customers.json"
```

```python
# parse the Json Line text format file line by line
cust = []
with open(path) as f:
    for line in f:
        cust.append(json.loads(line))
```

```python
# flatten the json into a DataFrame
df_cust = pd.json_normalize(cust)
df_cust.head()
```

```python
def create_agg_feature(row, col, feat, agg):
    """
    Apply aggregations to nested columns.
    
    :param row: row on which to apply the aggregation.
    :param col: column that contains nested json
    :param feat: the feature of interest within json
    :param agg: aggregation function to apply
    
    :return val: Aggregated value of selected feature
    """
    try:
        val = pd.json_normalize(row[col]).aggregate({feat: agg})
    except:
        val = 0
    return val
```

```python
# create order features for each customer
df_cust['maxOrderAmount'] = df_cust.apply(lambda row: create_agg_feature(row, col='orders',feat='orderAmount', agg="max"), axis=1)
df_cust['minOrderAmount'] = df_cust.apply(lambda row: create_agg_feature(row, col='orders', feat='orderAmount', agg="min"), axis=1)
df_cust['meanOrderAmount'] = df_cust.apply(lambda row: create_agg_feature(row, col='orders', feat='orderAmount', agg="mean"), axis=1)
df_cust['uniqueOrderShippingAddress'] = df_cust.apply(lambda row: create_agg_feature(row, col='orders', feat='orderShippingAddress', agg="nunique"), axis=1)
df_cust['totalOrders'] = df_cust.apply(lambda row: create_agg_feature(row, col='orders', feat='orderAmount', agg="count"), axis=1)

# create payment features for each customer
df_cust['paymentFailureRate'] = df_cust.apply(lambda row: create_agg_feature(row, col='paymentMethods', feat='paymentMethodRegistrationFailure', agg="mean"), axis=1)
df_cust['uniquePaymentMethodType'] = df_cust.apply(lambda row: create_agg_feature(row, col='paymentMethods', feat='paymentMethodType', agg="nunique"), axis=1)
df_cust['uniquePaymentMethodProvider'] = df_cust.apply(lambda row: create_agg_feature(row, col='paymentMethods', feat='paymentMethodProvider', agg="nunique"), axis=1)
df_cust['uniquePaymentMethodIssuer'] = df_cust.apply(lambda row: create_agg_feature(row, col='paymentMethods', feat='paymentMethodIssuer', agg="nunique"), axis=1)
df_cust['totalPaymentMethods'] = df_cust.apply(lambda row: create_agg_feature(row, col='paymentMethods', feat='paymentMethodId', agg="count"), axis=1) # total registered payment method types

# create transaction features for each customer
df_cust['transactionFailureRate'] = df_cust.apply(lambda row: create_agg_feature(row, col='transactions', feat='transactionFailed', agg="mean"), axis=1)
df_cust['maxTransactionAmount'] = df_cust.apply(lambda row: create_agg_feature(row, col='transactions', feat='transactionAmount', agg="max"), axis=1)
df_cust['minTransactionAmount'] = df_cust.apply(lambda row: create_agg_feature(row, col='transactions', feat='transactionAmount', agg="min"), axis=1)
df_cust['meanTransactionAmount'] = df_cust.apply(lambda row: create_agg_feature(row, col='transactions', feat='transactionAmount', agg="mean"), axis=1)
df_cust['totalTransactions'] = df_cust.apply(lambda row: create_agg_feature(row, col='transactions', feat='transactionId', agg="count"), axis=1) # total registered payment method types

# ratios
df_cust['PaymentTransactionRatio'] = df_cust['totalPaymentMethods'].div(df_cust['totalTransactions']).replace(np.inf, 0)

# email domain
# one-hot encoding large number of email domains relative to num obs not ideal, use target encoding
df_cust['emailDomain'] = df_cust.apply(lambda row: re.findall(r"[^@]*$", row['customer.customerEmail'])[0], axis=1)
df_cust['emailDomainGroup'] = [email if email in ['yahoo.com', 'gmail.com', 'hotmail.com']
                               else 'Other' for email in df_cust['emailDomain']]
```

Extract additional info from the nested jsons (orders, payments, transactions) such as payment types as proportion of total e.g. 75% of cust 1's payment types are Paypal.

```python
# add a customer ID to each record so unpacked arrays can be linked back to a customer
for i, record in enumerate(cust):
    cust[i].update({"id": i})
```

```python
# get nested order info associated with each customer
order_data = pd.json_normalize(data=cust, record_path="orders",
                                meta=["id"], errors='ignore')
order_data["count"] = 1

# count frequency of order states for each customer
order_states = order_data.pivot_table(values="count", index="id",
                                      columns="orderState", aggfunc="sum")\
                                      .fillna(0).add_prefix("orders_")

# proportion of orders that are failed/fulfilled/pending
df_orders = order_states[['orders_failed', 'orders_fulfilled', 'orders_pending']]\
                        .div(order_states.sum(axis=1), axis = 0)\
                        .astype(float)
```

```python
# get nested payment method type as proportion of total (e.g. higher prop of bitcoin may indicate fraud)
payment_data = pd.json_normalize(data=cust, record_path="paymentMethods",
                                meta=["id"], errors='ignore')
payment_data["count"] = 1

# count frequency of payment method types for each customer
df_payment = payment_data.pivot_table(values="count", index="id",
                                      columns="paymentMethodType", aggfunc="sum")\
                                      .fillna(0).add_prefix("payment_")


# proportion of payment types that are paypal/bitcoin/card etc.
df_payment = df_payment.div(df_payment.sum(axis=1), axis = 0)
df_payment.columns = [col.replace(" ", "") for col in df_payment.columns] # consistent col names
```

```python
dfs = [df_cust, df_orders, df_payment]

df_cust = reduce(lambda left,right: pd.merge(left, right, left_index=True, right_index=True, how='left'), dfs)

# where there are no orders, payments, transactions, fill with zeros
df_cust = df_cust.fillna(0)
```

```python
df = df_cust[['fraudulent', 'maxOrderAmount', 'minOrderAmount',
              'meanOrderAmount', 'uniqueOrderShippingAddress', 'totalOrders',
              'paymentFailureRate', 'uniquePaymentMethodType',
              'uniquePaymentMethodProvider', 'uniquePaymentMethodIssuer',
              'totalPaymentMethods', 'transactionFailureRate', 'maxTransactionAmount',
              'minTransactionAmount', 'meanTransactionAmount', 'totalTransactions',
              'PaymentTransactionRatio', 'emailDomainGroup',
              'orders_failed', 'orders_fulfilled', 'orders_pending',
              'payment_applepay', 'payment_bitcoin', 'payment_card',
              'payment_paypal']]
```

train/test split

```python
X = df.loc[:, df.columns != 'fraudulent']
y = df['fraudulent']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                    shuffle=True, random_state=101)
```

Apply Label Encoding to categorical variables

```python
def label_endode(X_train, X_test):
    """
    Apply label encoding to the categorical variables.
    
    :param X_train: DataFrame containing train set inputs.
    :param X_test: DataFrame containing test set inputs.
    :param variables: Categorical variables to which label
                      encoding will be applied.
    
    :returns X_train, X_test: train and test DataFrame
                              with label encoding applied.
    """
    # define categorical variables
    categorical_vars = list(X_train.select_dtypes(include="object").columns)
    
    label = LabelEncoder()
    
    X_train = X_train.copy()
    X_test = X_test.copy()
    
    X_train[categorical_vars] = label.fit_transform(X_train[categorical_vars].values)
    X_test[categorical_vars] = label.transform(X_test[categorical_vars].values)
    
    return X_train, X_test
```

```python
X_train, X_test = label_endode(X_train, X_test)
```

```python

xgb_cv = XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.01, gamma=0, 
                        min_child_weight=1, use_label_encoder=False,  eval_metric="logloss")
```

## Evaluation

Cross validation shows a mean test set accuracy of 75% but with a standard devistion of 11%, meaning performance in some folds was weaker than that achieved by a naive classifier whereby we predict the majority class.

```python
yhat = [0 for x in range(len(y_test))]
accuracy = accuracy_score(y_test, yhat)
print("Naive majority class prediction accuracy: {:.2%}".format(accuracy))
```

```python
scores = cross_val_score(xgb_cv, X_train, y_train, cv=10, scoring = "accuracy")
print("Accuracy Scores: \n", scores)
print("Mean:", scores.mean())
print("Standard Deviation:", scores.std())
```

```python
y_pred = cross_val_predict(xgb_cv, X_test, y_test, cv=4)
cm = confusion_matrix(y_test, y_pred)
cm_display = ConfusionMatrixDisplay(cm).plot()
```

Plot AUC

```python
xgb_cv.fit(X_train, y_train)
probs = xgb_cv.predict_proba(X_test)
y_score = pd.DataFrame(probs)[1]
fpr, tpr, _ = roc_curve(y_test, y_score)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```

## Feature Importance

```python
shap.initjs()
```

```python
shap_values = shap.TreeExplainer(xgb_cv).shap_values(X_train)
shap.summary_plot(shap_values, X_train, plot_type="bar")
```

- A summary plot of the SHAP values indicate that very high values of meanOrderAmount and PaymentTransactionRatio increase the risk of a customer being fraudulent, and these variables are the most predictive.

- A low proportion of paypal payments reduces the predicted fraudulence risk while medium to high levels increase the risk.

- There is one email domain that is most associated with lower risk

- Meanwhile a the email domain Yahoo.com (encoded with a '3', the highest value) reduces the predicted fraudlence risk.

- High rates of bitcoin usage (payment_bitcoin) and multiple unique payment method types (uniquePaymentMethodType) also increase the risk of fraud but are less predictive than aforementioned variables.

```python
shap.summary_plot(shap_values, features=X_train, feature_names=X_train.columns)
```

A closer look at the shap values using a dependence plot shows how an average Order Amount of above c.38 considerably increases the risk of fraudulence, and is correlated with the average number of transactoion amounts.

```python
inds = shap.approximate_interactions("meanOrderAmount", shap_values, X_train)
shap.dependence_plot("meanOrderAmount", shap_values, X_train, interaction_index="meanTransactionAmount")
```
